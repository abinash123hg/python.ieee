11. Define Features (X) and Target (y)

```python
X = data.drop('Mental_Health_Score', axis=1)
y = data['Mental_Health_Score']
```

12. Define the categorical columns

```python
categorical_cols = ['Gender','Academic_Level','Country','Most_Used_Platform','Relationship_Status']
```

13. Initialize and Fit OneHotEncoder

```python
encoder = OneHotEncoder(sparse_output=False, drop='first')
X_encoded = encoder.fit_transform(X[categorical_cols])
```

14. Encoded column names and create dataFrame

```python
encoded_cols = encoder.get_feature_names_out(categorical_cols)
X_encoded_df = pd.DataFrame(X_encoded, columns=encoded_cols)
```

15. Define Numerical Columns

```python
X_numerical = X[numerical_cols]
```

16. Extract Numerical Features

```python
X_numerical = X[numerical_cols]
```

17. Combine Encoded Categorical and Numerical Features

```python
X_final = pd.concat([X_numerical.reset_index(drop=True),
X_encoded_df.reset_index(drop=True)], axis=1)
```

18. Split Data into Train and Test Sets

```python
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)
```

19. Initialize and Train Random Forest Model

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

20. Predictions on test set

```python
y_pred = model.predict(X_test)
```

21. Calculate Metrics

```python
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print("\nmodel results:")
print(f"R² Score: {r2:f} ")
print(f"MAE: {mae:f} ")
```

```python
importances = model.feature_importances_
feat_df = pd.DataFrame({'Feature': X_final.columns, 'Importance': importances}).sort_values(by='Importance', ascending=False).head(5)
print(feat_df)
```

“These are the top 5 predictors of mental health scores.”

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(x='Importance', y='Feature', data=feat_df)
plt.title('Top 5 Features Affecting Mental Health Score')
plt.tight_layout()
plt.show()
```

"Usage hours and sleep are the main factors that predict mental health scores."

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(35, 25)) 
plot_tree(model.estimators_[0],feature_names=X_final.columns,filled=True, rounded=True, max_depth=4,fontsize=14,proportion=True)
plt.show()
```

“This decision tree shows how the model uses important features such as usage hours and sleep to predict mental health scores.”
