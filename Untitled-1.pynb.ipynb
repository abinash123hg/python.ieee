```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.tree import plot_tree
# Load the dataset (update path if needed)
data = pd.read_csv(r"C:\Users\VICTUS\Downloads\Students Social Media Addiction (2) (1).csv")
# Drop columns we don't need
data = data.drop(['Student_ID', 'Affects_Academic_Performance', 
                  'Conflicts_Over_Social_Media', 'Addicted_Score'], axis=1)
# Show first few rows to check data
print("First 5 Rows of Data:")
print(data.head())
# Features (X) and target (y = Mental Health Score)
X = data.drop('Mental_Health_Score', axis=1)
y = data['Mental_Health_Score']
# Convert categorical columns (Gender, Country, etc.) to numbers
categorical_cols = ['Gender', 'Academic_Level', 'Country', 'Most_Used_Platform', 'Relationship_Status']
encoder = OneHotEncoder(sparse_output=False, drop='first')  # Simple encoding, drop first to avoid dummy trap
X_encoded = encoder.fit_transform(X[categorical_cols])
encoded_cols = encoder.get_feature_names_out(categorical_cols)
X_encoded_df = pd.DataFrame(X_encoded, columns=encoded_cols)
# Combine encoded categorical with numerical columns
numerical_cols = ['Age', 'Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night']
X_numerical = X[numerical_cols]
X_final = pd.concat([X_numerical.reset_index(drop=True), X_encoded_df.reset_index(drop=True)], axis=1)
# Split data: 80% train, 20% test
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42)
# Train Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
# Make predictions
y_pred = model.predict(X_test)
# Calculate metrics
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print("\nModel Results:")
print(f"RÂ² Score: {r2:.4f} (how well model fits data)")
print(f"MAE: {mae:.4f} (average prediction error)")
# Save metrics for paper
results = pd.DataFrame({
    'Model': ['Random Forest'],
    'RÂ²': [r2],
    'MAE': [mae]
})
results.to_csv('model_results.csv', index=False)
print("\nResults saved to 'model_results.csv'")
# Feature importance plot
feature_names = X_final.columns
importances = model.feature_importances_
feat_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances}).sort_values('Importance', ascending=False).head(5)
plt.figure(figsize=(8, 5))
sns.barplot(x='Importance', y='Feature', data=feat_df)
plt.title('Top 5 Features Affecting Mental Health Score')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.savefig('feature_importance.png')
plt.show()
# Plot one decision tree (first tree from Random Forest)
plt.figure(figsize=(15, 10))
plot_tree(model.estimators_[0], feature_names=feature_names, filled=True, rounded=True, max_depth8=5)
plt.title('Decision Tree from Random Forest Model')
plt.savefig('decision_tree.png')
plt.show()
# Sample predictions for presentation
sample_df = pd.DataFrame({
    'Actual Mental Health Score': y_test[:5].values,
    'Predicted Score': y_pred[:5].round(2)
})
print("\nSample Predictions:")
print(sample_df)
sample_df.to_csv('sample_predictions.csv', index=False)
```
